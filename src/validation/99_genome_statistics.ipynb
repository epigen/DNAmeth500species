```r
latin_names <- list("vicPac2" = "Vicugna pacos", "dasNov3" = "Dasypus novemcinctus", "gasAcu1" = "Gasterosteus aculeatus", "calJac4" = "Callithrix jacchus", "choHof1" = "Choloepus hoffmanni",
"aptMan1" = "Apteryx australis", "dipOrd1" = "Dipodomys ordii", "canFam5" = "Canis lupus familiaris", "equCab3" = "Equus caballus", "cavPor3" = "Cavia porcellus", "mm39" = "Mus musculus",
"loxAfr3" = " Loxodonta africana", "neoSch1" = "Monachus schauinslandi", "eriEur2" = "Erinaceus europaeus", "bisBis1" = "Bison bison bison", "criGri1" = "Cricetulus griseus",
"micMur2" = "Microcebus murinus", "rn7" = "Rattus norvegicus", "papAnu4" = "Papio anubis", "strPur2" = "Strongylocentrotus purpuratus", "ci3" = "C. intestinalis",
"oryLat2" = "Oryzias latipes", "petMar3" = "Petromyzon marinus", "tetNig2" = "Tetraodon nigroviridis", "melGal5" = "Meleagris gallopavo","triMan1" = " Trichechus manatus latirostris",
"proCap1" = "Procavia capensis", "echTel2" = "Echinops telfairi", "macFas5" = "Macaca fascicularis",  "rhiRox1" = "Rhinopithecus roxellana", "nasLar1" = "Nasalis larvatus")
```

```r
genomes_df[is.na(V4)]$V4 <- as.character(latin_names[genomes_df[is.na(V4)]$V1])
```

```r
genome_summary = data.table()
```

```r
for(i in seq_along(genomes_df$V1)){
    
genome_id <- genomes_df[i,]$V1

kmer01 <- as.data.frame(t(read.csv(file.path(genome_dir, genome_id,"kmer01"), row.names = 1, sep = " ")))["order",]                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                    
kmer01$genome <- genome_id

kmer02 <- as.data.frame(t(read.csv(file.path(genome_dir, genome_id,"kmer02"), row.names = 1, sep = " "))["order",c("CG"), drop = FALSE])
kmer02$genome <- genome_id

kmer <- full_join(kmer01, kmer02)

kmer$CG_OE <- kmer$CG/(kmer$C*kmer$G)
kmer$CG_freq <- kmer$C + kmer$G

genome_size <- fread(file.path(genome_dir,genome_id, paste0(genome_id, ".chrom.sizes")))

kmer$total_bp <- sum(genome_size$V2)
kmer$N_chr <- NROW(genome_size)

annot_files <- list.files(file.path(genome_dir, genome_id, "tracks"))
repeat_file <- file.path(genome_dir, genome_id, "tracks", paste0(genome_id, "_rmsk.bed"))
if(!file.exists(repeat_file)){
    repeat_file <- file.path(genome_dir, genome_id, "tracks", paste0(genome_id, "_simpleRepeat.bed"))
    if(!file.exists(repeat_file)){
        
        read_repeat_flag <- FALSE
    }else{
        read_repeat_flag <- TRUE
    }
}else read_repeat_flag <- TRUE
       
if(read_repeat_flag){
    repeats <- fread(repeat_file)

repeats[, width:= V3-V2, ]

kmer$repeats_ratio <- sum(repeats$width)/kmer$total_bp
}else kmer$repeats_ratio <- NA



ref_file <- annot_files[grepl("ref", tolower(annot_files))]

gene_annot <- fread(file.path(genome_dir, genome_id, "tracks",ref_file))

kmer$n_genes_annotated <- NROW(gene_annot)

kmer$gene_annotation <- gsub(paste0(genome_id, "_"), "", gsub(".bed", "", ref_file))
   kmer$common_name <- genomes_df[i,]$V2
kmer$latin_name <- genomes_df[i,]$V4 
genome_summary <- rbind(kmer, genome_summary)


}
```
